```python
# Import dataset
meteo = pd.read_csv('meteogram_whistler.csv', parse_dates=['datetime']) 
meteo.head()
```

```python
precip = alt.Chart(meteo).mark_bar(color='green', size=16).encode(
    alt.X('datetime:T', title=None, axis=alt.Axis(labelAngle=90)),
    alt.Y('pc_3hr:Q', title = 'Precip-3hr (mm)')
).properties(width=800, height=150)


precip = precip + rules
precip
```

```python
temp_line = alt.Chart(meteo).mark_line(color='red', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('Ta', title='T/Td@2m (C)')
).properties(width=800, height=150)

temp = alt.Chart(meteo).mark_line(color='red', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('Ta', title='T/Td@2m (C)')
).properties(width=800, height=150)

zero = alt.Chart(meteo).mark_line(color='green', strokeWidth=1.5, opacity = 0.6).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('zero', title='T/Td@2m (C)')
).properties(width=800, height=150)

dew = alt.Chart(meteo).mark_line(color='blue', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('Td', title='T/Td@2m (C)')
).properties(width=800, height=150)

temp_circle = alt.Chart(meteo.dropna()).mark_point(color='red', size=20).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('Ta_3hr', title='T/Td@2m (C)')
).properties(width=800, height=150)

dew_circle = alt.Chart(meteo.dropna()).mark_point(color='blue', shape="triangle", size=20).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('Td_3h', title='T/Td@2m (C)')
).properties(width=800, height=150)

area = alt.Chart(meteo).mark_area(color='#F8D223', opacity=0.5).encode(
    alt.X('datetime:T'),
    alt.Y('Ta:Q'),
    alt.Y2('Td:Q')
).properties(width=800, height=150)

temp_chart = zero + dew + dew_circle +temp_line + temp_circle + area + rules

temp_chart
```

```python
solar = alt.Chart(meteo).mark_line(color='red', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('solar', title='SWR(Wm^2) (red)')
).properties(width=800, height=150)

cloud_line = alt.Chart(meteo).mark_line(color='blue', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('cloud', title='Cloud Cover (%)')
).properties(width=800, height=150)

cloud_area = alt.Chart(meteo).mark_area(color='#B4A667', opacity=0.5).encode(
    alt.X('datetime:T'),
    alt.Y('cloud:Q')
).properties(width=800, height=150)

cloud = cloud_area + cloud_line + rules
    
cloud_solar = alt.layer(cloud, solar).resolve_scale(
    y='independent'
)


cloud_solar
```

```python
wind_3k = alt.Chart(meteo).mark_line(color='red', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('wspd_3k', title='Wind(kph) @2/3km(green/red)')
).properties(width=800, height=150)

wind_2k = alt.Chart(meteo).mark_line(color='green', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('wspd_2k', title='Wind(kph) @2/3km(green/red)')
).properties(width=800, height=150)

text1 = alt.Chart(meteo.dropna()).mark_text(color='green', size=10).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('wspd_2k', title='Wind(kph) @2/3km(green/red)'),
    text="wdir_2k:N",
).properties(width=800, height=150)

text2 = alt.Chart(meteo.dropna()).mark_text(color='red', size=10).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('wspd_3k', title='Wind(kph) @2/3km(green/red)'),
    text=alt.Text("wdir_3k:N")
).properties(width=800, height=150)

wind_2plot = wind_2k + text1
wind_3plot = wind_3k + text2

mslp = alt.Chart(meteo).mark_line(color='blue', strokeWidth=1.5).encode(
    alt.X('datetime:T',  title=None,),
    alt.Y('mslp', title='MSLP (mb) (blue)', scale=alt.Scale(domain=[996,1024]))
).properties(width=800, height=150)


wind = wind_2plot + wind_3plot + rules

wind_mslp = alt.layer(wind, mslp).resolve_scale(
    y='independent'
)


wind_mslp
```

```python

time_height_a = alt.Chart(meteo).mark_line(color='#F8D223', strokeWidth=2).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('height_a', title='Temp (C) & RH (%)', scale=alt.Scale(domain=[0,5000]))
).properties(width=800, height=400)

time_height_b = alt.Chart(meteo).mark_line(color='#F8D223', strokeWidth=2, strokeDash=[6, 4]).encode(
    alt.X('datetime:T',  title=None),
    alt.Y('height_b', title='Temp (C) & RH (%)', scale=alt.Scale(domain=[0,5000]))
).properties(width=800, height=400)


top_panel = time_height_a + time_height_b + rules.properties(title='Whistler @ Sea to Sky (122.96W, 50.06N, 2178/1202m) 20171120 GEM-glb')
top_panel
```

```python
rule = alt.Chart(pd.DataFrame({
  'datetime': [('2017-11-21 04:00:00')],
  'color': ['black']
})).mark_rule(opacity=0.6, strokeDash=[3,3]).encode(
  x='datetime:T',
  color=alt.Color('color:N', scale=None)
)

rule2 = alt.Chart(pd.DataFrame({
  'datetime': [('2017-11-22 04:00:00')],
  'color': ['black']
})).mark_rule(opacity=0.6, strokeDash=[3,3]).encode(
  x='datetime:T',
  color=alt.Color('color:N', scale=None)
)

rule3 = alt.Chart(pd.DataFrame({
  'datetime': [('2017-11-23 04:00:00')],
  'color': ['black']
})).mark_rule(opacity=0.6, strokeDash=[3,3]).encode(
  x='datetime:T',
  color=alt.Color('color:N', scale=None)
)

rule4 = alt.Chart(pd.DataFrame({
  'datetime': [('2017-11-24 04:00:00')],
  'color': ['black']
})).mark_rule(opacity=0.6, strokeDash=[3,3]).encode(
  x='datetime:T',
  color=alt.Color('color:N', scale=None)
)

rule5 = alt.Chart(pd.DataFrame({
  'datetime': [('2017-11-25 04:00:00')],
  'color': ['black']
})).mark_rule(opacity=0.6, strokeDash=[3,3]).encode(
  x='datetime:T',
  color=alt.Color('color:N', scale=None)
)

rules = rule + rule2 + rule3 + rule4 + rule5
rules
```
