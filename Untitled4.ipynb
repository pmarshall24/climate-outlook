```python
import os
import shutil
import exifread
from datetime import datetime
```

```python
# Path to the folder containing the images
folder_path = 'C://Users/pmarshal/Pictures/seymour_river2'


# Function to delete every other photo
def delete_every_other_photo(folder_path):
    delete_flag = False
    for filename in sorted(os.listdir(folder_path)):  # Sort the filenames to ensure consistent order
        if filename.endswith(".jpg"):  # Adjust file extension as needed
            file_path = os.path.join(folder_path, filename)
            if delete_flag:
                os.remove(file_path)
                print(f"Deleted {filename}")
            delete_flag = not delete_flag

# Call the function
delete_every_other_photo(folder_path)
```
