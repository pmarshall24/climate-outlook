```python
# Replace 'your_data.csv' with the actual file path
monthly_q = pd.read_csv('cap_lakehead_monthlyQ.csv')
monthly_q.head()
```

```python
# Compute the average flow for Jun, Jul, Aug, Sep
monthly_q['summer_avg'] = monthly_q[['jun', 'jul', 'aug', 'sep']].mean(axis=1)
monthly_q.head()
```

```python
# Calculate 10-year moving average
monthly_q['moving_avg_10yr'] = monthly_q['summer_avg'].rolling(window=10).mean()
monthly_q.head(12)
monthly_q.to_csv('monthly_q.csv', index=False)
```

```python
# Prepare the data for Altair
summer_data = monthly_q[['year', 'summer_avg', 'moving_avg_10yr']]

# Create a bar chart
highlight_years = [1915, 1926, 1944, 1985, 2015]

# Construct the condition for highlighting
highlight_condition = (
    (alt.datum.year == 1915) | 
    (alt.datum.year == 1926) | 
    (alt.datum.year == 1944) | 
    (alt.datum.year == 1985) | 
    (alt.datum.year == 2015)
)

chart = alt.Chart(summer_data).mark_bar().encode(
    x=alt.X('year:O', title=None, axis=alt.Axis(labelExpr="datum.value % 5 == 0 ? datum.value : ''")),
    y=alt.Y('summer_avg:Q', title='Average Summer Inflow (m^3/s)'),
    color=alt.condition(
        highlight_condition,  # Use the constructed highlight condition
        alt.value('darkblue'),     # Highlight with red
        alt.value('darkgrey') # Grey for the rest
    ),
    tooltip=['year', 'summer_avg']
).properties(
    width=800,
    height=200
)

chart
```

```python
# Create a line chart for the moving average
moving_avg_line = alt.Chart(summer_data).mark_line(color='black', opacity=0.6).encode(
    x=alt.X('year:O', title=None),
    y=alt.Y('moving_avg_10yr:Q', title='Average Summer Inflow (m^3/s)')
)
moving_avg_line
```
