```python
precip_flow = pd.read_csv("precip_flows.csv")
precip_flow['date'] = pd.to_datetime(precip_flow['date'])
precip_flow['year'] = pd.DatetimeIndex(precip_flow['date']).year
precip_flow['month'] = pd.DatetimeIndex(precip_flow['date']).month
precip_flow['day'] = pd.DatetimeIndex(precip_flow['date']).day
precip_flow['DOY'] = pd.DatetimeIndex(precip_flow['date']).dayofyear
precip_flow.head()
```

```python
cap_precip_bar = alt.Chart(precip_flow).mark_bar(color='darkgreen').encode(
    alt.X('monthdate(date):T'),
    alt.Y('cap_precip:Q')
).properties(width=600)

cap_flow_line = alt.Chart(precip_flow).mark_line(stroke='#5276A7', interpolate='monotone').encode(
    alt.X('monthdate(date):T'),
    alt.Y('cap_flow:Q')
).properties(width=600)

#alt.layer(cap_precip_bar, cap_flow_line).resolve_scale(
#    y='independent'
#)
cap_precip_bar
```

```python
cap_precip_bar = alt.Chart(precip_flow).mark_bar(color='darkgreen', width=2, opacity=0.4).encode(
    alt.X('monthdate(date):T', title=None),
    alt.Y('cap_precip:Q', title='Daily Precipitation (mm)', scale=alt.Scale(domain=[0, 250]))
).properties(width=600, height=250)

# Define the second chart with Y-axis going from top down
cap_flow_line = alt.Chart(precip_flow).mark_line(stroke='#5276A7', interpolate='monotone').encode(
    alt.X('monthdate(date):T', title=None),
    alt.Y('cap_flow:Q', title='Mean Daily Inflow (m/s)', scale=alt.Scale(domain=[0, 250]))
).properties(width=600, height=250)

# Concatenate the charts with a shared X-axis
layered_chart = alt.layer(cap_flow_line, cap_precip_bar).resolve_scale(y='independent')

layered_chart.properties(title='Capilano Precipitation and Inflows')
```

```python
sey_precip_bar = alt.Chart(precip_flow).mark_bar(color='darkgreen', width=2, opacity=0.4).encode(
    alt.X('monthdate(date):T', title=None),
    alt.Y('sey_precip:Q', title='Daily Precipitation (mm)', scale=alt.Scale(domain=[0, 100]))
).properties(width=600, height=250)

# Define the second chart with Y-axis going from top down
sey_flow_line = alt.Chart(precip_flow).mark_line(stroke='#5276A7', interpolate='monotone').encode(
    alt.X('monthdate(date):T', title=None),
    alt.Y('sey_flow:Q', title='Mean Daily Inflow (m/s)', scale=alt.Scale(domain=[0, 100]))
).properties(width=600, height=250)

# Concatenate the charts with a shared X-axis
layered_chart2 = alt.layer(sey_flow_line, sey_precip_bar).resolve_scale(y='independent')

layered_chart2.properties(title='Seymour Precipitation and Inflows')
```
