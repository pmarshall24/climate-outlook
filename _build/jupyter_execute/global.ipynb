Many parts of the world have experienced intense and prolonged heat waves in 2023. Globally, June was the [hottest ever](https://www.noaa.gov/news/earth-just-had-its-hottest-june-on-record) this year. Moving into July, headlines were made when average global temperatures climbed into all-time record territory and stayed there all month. This July was the [hottest month ever on record](https://public.wmo.int/en/media/press-release/july-2023-set-be-hottest-month-record). Temperatures have remained hot in August with several parts of the world suffering intense heatwaves (South America, southern Europe, southern US).

The chart below shows the global average temperature from 1979 to present. The bold black line represents 2023, while the orange line highlights 2022. Prior to 2023, the record global high temperature was 19.92 C, set on August 14, 2016. It's still quite early for annual estimates, but it looks very likely that 2023 will be the hottest year ever.

As an example, hot waters in the Gulf of Mexico caused Hurricane Idalia to rapidly intensify into a category 3 hurricane as it made landfall on the Florida coast on August 29.

### World Extreme Weather Events in 2023

Below is a map highlighting some of the most extreme weather events so far this year. This includes extreme heatwaves, powerful storms, flooding, and large intense wildfires. Combined, these events have resulted in thousands of lives lost and homes destroyed, and have cost billions in response and recovery. 

August alone was packed with devastating weather events. South America experienced extreme mid-winter heat with temperatures 20-25 degrees warmer than average. This is considered to be one of the **most extreme weather events ever**. Heat and wildfires simultaneously struck North America and Europe. Greece is currently battling the **largest wildfire ever in Europe**. The town of Lahaina on Maui was almost completely destroyed on August 8th. 115 people were killed in this fire, with hundreds still missing. Finally, category 3 Hurricane Idalia struck Florida's Big Bend region on August 29. This storm brought catastrophic storm surge and destructive winds, and is considered to be the **strongest storm to strike this part of Florida in the past 130 years**. 

You can hover over the icons on the map to see the name and date of the event, and click on the icons for more information.
