### Air Temperatures
```{admonition} **Off the Charts Heat** 
:class: danger
Summer 2023 was the hottest in history with exceptionally warm global temperatures since June
```

Heat has grabbed headlines around the world in 2023. Many parts of the world have experienced intense and prolonged heat waves so far this year. NASA has announced that [summer 2023 was the hottest since records began](https://www.nasa.gov/press-release/nasa-announces-summer-2023-hottest-on-record) in the late 1800â€™s. Global temperatures have been consistently in record-breaking territory since June. July was the [hottest month ever on recorded history](https://public.wmo.int/en/media/press-release/july-2023-set-be-hottest-month-record), and the temperature anomaly has actually increased moving into September. These warm temperatures have had led to massive and deadly wildfires, a dramatic loss of Arctic and Antarctic sea ice, and has fueled intense storm systems around the world.   

The chart below shows the global average temperature from 1979 to present. The bold black line represents 2023, while the orange line highlights 2022. Prior to 2023, the record global high temperature was 16.92 C, set on August 14, 2016. The new hottest day on record is July 6, 2023 with a global temperature of 17.23 C. It looks almost certain that 2023 will set a new annual global heat record.

### Sea Surface Temperatures
Sea surface temperatures have been in completely new territory for most of 2023. The chart below is absolutely shocking! This means there is significant heat (energy) at the surface of the ocean to interact with the atmosphere. As a result, we are seeing hotter average air temperatures and more intense storms. Hot waters in the Atlantic and Gulf of Mexico have fueled hurricanes, causing some to intensify rapidly and become monster storms (E.g. Hurricane Lee).

### Sea Ice Extent
Antarctic sea ice extent has been well below the previous record-low level throughout 2023. The National Snow and Ice Data Center (NSIDC) just announced that the [2023 maximum Antarctic sea ice extent is the lowest observed](https://nsidc.org/news-analyses/news-stories/antarctic-sea-ice-hits-record-low-maximum-extent-2023)(preliminary report). The ice extent is substantially lower than the previous mark set in 1986. Arctic sea ice extent is also below normal this summer. Again, the chart below is very alarming.

### World Extreme Weather Events in 2023

Below is a map highlighting some of the most extreme weather events so far this year. This includes extreme heatwaves, powerful storms, flooding, and large intense wildfires. Combined, these events have resulted in thousands of lives lost and homes destroyed, and have cost billions in response and recovery. 

August alone was packed with devastating weather events. South America experienced extreme mid-winter heat with temperatures 20-25 degrees warmer than average. This is considered to be one of the **most extreme weather events ever**. Heat and wildfires simultaneously struck North America and Europe. Greece is currently battling the **largest wildfire ever in Europe**. The town of Lahaina on Maui was almost completely destroyed on August 8th. 115 people were killed in this fire, with hundreds still missing. Finally, category 3 Hurricane Idalia struck Florida's Big Bend region on August 29. This storm brought catastrophic storm surge and destructive winds, and is considered to be the **strongest storm to strike this part of Florida in the past 130 years**.
