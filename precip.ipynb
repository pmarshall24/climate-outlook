# Precipitation
The charts below show precipitation trends in the watersheds. You will see plots with annual accumulated precipitation, and other showing daily precipitation amounts. 

```python
# Import dataset
qd_pc = pd.read_csv('data/QD1_temps.csv', parse_dates=['date']) 
qd_pc['wtr_day_yr'] = pd.to_datetime('2023-01-01') + pd.to_timedelta(qd_pc['wtr_yr_day'] - 93, unit='D')
qd_pc = qd_pc.dropna()
qd_pc['year'] = pd.DatetimeIndex(qd_pc['date']).year
qd_pc['month'] = pd.DatetimeIndex(qd_pc['date']).month
qd_pc['day'] = pd.DatetimeIndex(qd_pc['date']).day
qd_pc['DOY'] = pd.DatetimeIndex(qd_pc['date']).dayofyear
qd_pc['quarter'] = pd.DatetimeIndex(qd_pc['date']).quarter
qd_pc.head()
```

```python
#Caluclate statistics
qd_pc['mean'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform('mean').round(2)
qd_pc['max'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform('max').round(2)
qd_pc['min'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform('min').round(2)
qd_pc['std'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform('std').round(2)
qd_pc['sem'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform('sem').round(2)
qd_pc['ci95_hi'] = qd_pc['mean'] + 1.96* qd_pc['sem']
qd_pc['ci95_lo'] = qd_pc['mean'] - 1.96* qd_pc['sem']
qd_pc['ci95_hi'] = qd_pc['ci95_hi'].round(2)
qd_pc['ci95_lo'] = qd_pc['ci95_lo'].round(2)
qd_pc['percentile_5th'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform(lambda x: np.percentile(x, 5))
qd_pc['percentile_95th'] = qd_pc.groupby('wtr_yr_day')['rain_cum_wyr'].transform(lambda x: np.percentile(x, 95))
qd_pc.tail()
```

```python
qd_pc = qd_pc.drop(['mean_TA', 'min_TA', 'max_TA'], axis=1)
```

```python
last_few = qd_pc.loc[qd_pc['wtr_yr'] >= 2022]
last_few.tail()
# Export DataFrame to CSV
#last_few.to_csv('output.csv', index=False)
```

```python
title_plot = alt.TitleParams(
   text='Cummulative Precipitation - Lower Capilano Watershed',
   subtitle="Gray area is 5th to 95th percentile. Red dashed line is 20-year mean",
   anchor='middle',
   fontSize=14,
   fontWeight='bold')

pc_mean = alt.Chart(qd_pc[qd_pc['wtr_yr'] == 2022]).mark_line(color= "darkred", size= 1.5, strokeDash=[4, 2]).encode(
    alt.X('wtr_day_yr:T', title=None,
          axis=alt.Axis(format='%b-%d', tickCount=12, labelOverlap='parity')),
    alt.Y('mean:Q')
).properties(width=600)

area = alt.Chart(qd_pc[qd_pc['wtr_yr'] == 2022]).mark_area(color='lightgray', opacity=0.5).encode(
    alt.X('wtr_day_yr:T', title=None,
          axis=alt.Axis(format='%b-%d', tickCount=12, labelOverlap='parity')),
    alt.Y('percentile_95th:Q', title=''),
    alt.Y2('percentile_5th:Q')
).properties(width=600, height=300)

past_years = alt.Chart(last_few, title=title_plot).mark_line(size= 1.5, opacity=0.8).encode(
    alt.X('wtr_day_yr:T', title=None,
          axis=alt.Axis(format='%b-%d', tickCount=12, labelOverlap='parity')),
    alt.Y('rain_cum_wyr:Q', title='Cumulative Precipitation (mm)'),
    color=alt.Color('wtr_yr:O', title='Water Year', scale=alt.Scale(scheme='bluepurple')),
    size=alt.condition(
        alt.FieldEqualPredicate(field='wtr_yr', equal=2025),  # If year is 2025
        alt.value(3),  # Thicker line for 2025
        alt.value(1.5)  # Default thickness for other years
    )
).properties(width=600)

total = area + pc_mean + past_years
total
```

### Annual Accumulated Precipitation 
The following charts show accumulated annual precipitation. The first chart displays accumulated precipitation over the water year (Oct 1 - Sep 30), and the second chart shows Jan 1 to Dec 31. You can select previous years to see how this year compares.  

```python
alt.data_transformers.disable_max_rows()

title2 = alt.TitleParams(
   text='Lower Capilano Watershed',
   subtitle="Cummulative Precipitation - Water Year (2003-2024)",
   anchor='middle',
   fontSize=14,
   fontWeight='bold')
# Create a selection for the year
#wtr_year_selection = alt.selection_single(
#    name='Select',
#    fields=['wtr_yr'],
#    bind=alt.binding_select(options=list(qd_pc['date'].dt.year.unique()), name='Select Year '),
#    init={'wtr_yr': 2024}  # Initial selected year
#)

wtr_year_selection = alt.selection_single(
    name='Select',
    fields=['wtr_yr'],
    bind=alt.binding_select(options=list(qd_pc['wtr_yr'].unique()), name='Select Year '),
    init={'wtr_yr': 2024}  # Initial selected year
)

# Create a line chart using Altair
chart2 = alt.Chart(qd_pc, title=title2).mark_line().encode(
    alt.X('wtr_yr_day', title ='Water Year - October 1 (0) to September 30 (365)', scale=alt.Scale(domain=[0,365], nice=False), axis=alt.Axis(values=[1,32,62,93,124,152,183,213,244,274,305,336])),
    alt.Y('rain_cum_wyr', title='Accumulated Precipitation (mm)'),
    color=alt.condition(wtr_year_selection, 'wtr_yr:N', alt.value('gray'), title='Water Year'),
    strokeWidth=alt.condition(wtr_year_selection, alt.value(2.5), alt.value(1)),
    opacity=alt.condition(wtr_year_selection, alt.value(1.0), alt.value(0.6))
).add_selection(
    wtr_year_selection
).properties(
    width=600
)

chart2
```

```python
title3 = alt.TitleParams(
   text='Lower Capilano Watershed',
   subtitle="Annual Cummulative Precipitation (2003-2023)",
   anchor='middle',
   fontSize=14,
   fontWeight='bold')
# Create a selection for the year
year_selection = alt.selection_single(
    name='Select',
    fields=['year'],
    bind=alt.binding_select(options=list(qd_pc['date'].dt.year.unique()), name='Select Year '),
    init={'year': 2023}  # Initial selected year
)

# Create a line chart using Altair
chart = alt.Chart(qd_pc).mark_line().encode(
    alt.X('monthdate(date):T', title =None),
    alt.Y('rain_cum_yr:Q', title='Accumulated Precipitation (mm)'),
    color=alt.condition(year_selection, 'year:N', alt.value('gray'), title='Year', legend=None),
    strokeWidth=alt.condition(year_selection, alt.value(2.5), alt.value(1)),
    opacity=alt.condition(year_selection, alt.value(1.0), alt.value(0.6))
).add_selection(
    year_selection
).properties(
    width=600
)

# Create a line chart using Altair
mean_chart = alt.Chart(qd_pc).mark_line(color='black', opacity=0.6, strokeDash=[4, 2], size= 2).encode(
    alt.X('monthdate(date):T', title =None),
    alt.Y('mean(rain_cum_yr):Q', title='Accumulated Precipitation (mm)')
).properties(
    width=600
)

chart + mean_chart
```
